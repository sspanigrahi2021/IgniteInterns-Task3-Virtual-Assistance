##### Rule-Based Virtual Assistant Implementation

```python
class RuleBasedAssistant:
    def __init__(self):
        self.commands = {
            "hello": self.greet,
            "help": self.help,
            "set reminder": self.set_reminder,
            "weather": self.weather_update
        }

    def greet(self):
        return "Hello! How can I assist you today?"

    def help(self):
        return "I can assist you with the following tasks:\n- Answering questions\n- Setting reminders\n- Providing weather updates"

    def set_reminder(self):
        return "Reminder set successfully."

    def weather_update(self):
        return "The current weather is sunny with a temperature of 25°C."

    def respond(self, command):
        for key in self.commands:
            if key in command.lower():
                return self.commands[key]()
        return "I'm sorry, I don't understand that command."

# Example usage
assistant = RuleBasedAssistant()
print(assistant.respond("hello"))
print(assistant.respond("help"))
print(assistant.respond("set reminder"))
print(assistant.respond("weather"))
```

#### NLU and Sentiment Analysis

```python
import spacy
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import speech_recognition as sr

# Load spaCy model for NLU
nlp = spacy.load("en_core_web_sm")

# Initialize sentiment analyzer
sentiment_analyzer = SentimentIntensityAnalyzer()

class AIAssistant(RuleBasedAssistant):
    def __init__(self):
        super().__init__()

    def analyze_sentiment(self, text):
        sentiment = sentiment_analyzer.polarity_scores(text)
        return sentiment

    def process_speech(self):
        recognizer = sr.Recognizer()
        with sr.Microphone() as source:
            print("Listening...")
            audio = recognizer.listen(source)
            try:
                text = recognizer.recognize_google(audio)
                print(f"You said: {text}")
                return text
            except sr.UnknownValueError:
                return "Sorry, I did not understand that."
            except sr.RequestError:
                return "Sorry, my speech service is down."

    def respond(self, command):
        doc = nlp(command)
        entities = {ent.label_: ent.text for ent in doc.ents}
        sentiment = self.analyze_sentiment(command)
        print(f"Entities: {entities}")
        print(f"Sentiment: {sentiment}")
        return super().respond(command)

# Example usage
ai_assistant = AIAssistant()
command = ai_assistant.process_speech()
print(ai_assistant.respond(command))
```

#### Email Sending Functionality

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

class AIAssistantWithEmail(AIAssistant):
    def __init__(self):
        super().__init__()

    def send_email(self, to_address, subject, body):
        # Replace with your email credentials
        sender_email = "sspanigrahi@gmail.com"
        sender_password = "#Shiva2001"

        # Compose the email message
        msg = MIMEMultipart()
        msg['From'] = sender_email
        msg['To'] = to_address
        msg['Subject'] = subject
        msg.attach(MIMEText(body, 'plain'))

        try:
            # Connect to the SMTP server
            server = smtplib.SMTP('smtp.gmail.com', 587)
            server.starttls()
            server.login(sender_email, sender_password)
            text = msg.as_string()
            # Send the email
            server.sendmail(sender_email, to_address, text)
            server.quit()
            return "Email sent successfully."
        except Exception as e:
            return f"Failed to send email. Error: {str(e)}"

# Example usage
assistant_with_email = AIAssistantWithEmail()
to_address = "recipient@example.com"
subject = "Test Email"
body = "This is a test email sent from the virtual assistant."
print(assistant_with_email.send_email(to_address, subject, body))
```

```python
from newsapi import NewsApiClient

class AIAssistantWithNews(AIAssistant):
    def __init__(self):
        super().__init__()
        self.news_api = NewsApiClient(api_key='your_news_api_key')

    def get_top_headlines(self):
        try:
            top_headlines = self.news_api.get_top_headlines(language='en', page_size=5)
            articles = top_headlines['articles']
            headlines = [article['title'] for article in articles]
            return headlines
        except Exception as e:
            return f"Failed to fetch news headlines. Error: {str(e)}"

# Example usage
assistant_with_news = AIAssistantWithNews()
headlines = assistant_with_news.get_top_headlines()
print("Top News Headlines:")
for idx, headline in enumerate(headlines, start=1):
    print(f"{idx}. {headline}")
```
